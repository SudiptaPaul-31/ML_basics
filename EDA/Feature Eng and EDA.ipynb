## Algerian Forest Fires Dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
dataset = pd.read_csv('Algerian_forest_fires_dataset.csv', header = 0)
```

```python
dataset.head()
```

## Data Cleaning

```python
## missing values
dataset[dataset.isnull().any(axis=1)]
```

```python
dataset.loc[:122,"Region"]=0
dataset.loc[122:,"Region"]=1
df = dataset
```

```python
df.head()
```

```python
df.tail()
```

```python
df[['Region']]=df[['Region']].astype(int)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
## Removing the null values
df = df.dropna().reset_index(drop=True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]]
```

```python
## Dropping the 122nd row
df = df.drop(122).reset_index(drop=True)
```

```python
df.iloc[[122]]
```

```python
df.columns
```

```python
## Fixing the empty spaces in between the columns
df.columns = df.columns.str.strip()
df.columns
```

```python
df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']] = df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']].astype(int)
```

```python
df.info()
```

### Changing the other columns to float data types

```python
objects = [features for features in df.columns if df[features].dtypes =='O']
```

```python
objects
```

```python
## Since classes should be either fire or not fire so we keep it object only
for i in objects:
    if i!= 'Classes':
        df[i] = df[i].astype(float)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.head()
```

```python
## Lets save the cleaned dataset to a new file
df.to_csv('Algerian_forest_fires_cleaned_dataset.csv', index=True)
```

## Exploratory Data Analysis

```python
## Drop day, month and year
df_copy = df.drop(['day','month','year'],axis = 1)
```

```python
df_copy.head()
```

```python
## Catergories in classes
df_copy['Classes'].value_counts()
```

```python
## Encoding of the categories in classes -> making the classes as 0 and 1's
df_copy['Classes'] = np.where(df_copy['Classes'].str.contains('not fire'),0,1)
```

```python
df_copy.head()
```

```python
df_copy.tail()
```

```python
df_copy['Classes'].value_counts()
```

```python
## Plot density plot for all features

sns.set(style="whitegrid")
df_copy.hist(bins=50,figsize=(20,15))
plt.show()
```

```python
## Percentage for the pie chart
percentage = df_copy['Classes'].value_counts(normalize=True)*100
```

```python
## Plotting Pie Chart
classlabels = ["Fire", "Not Fire"]
plt.figure(figsize=(12, 7))
plt.pie(percentage, labels=classlabels, autopct='%1.1f%%')
plt.title("Pie Chart of Classes")
plt.show()
```
