Algerian Forest Fires Dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
dataset = pd.read_csv('Algerian_forest_fires_dataset.csv')
```

```python
dataset.head()
```

```python
dataset.info()
```

## Data Cleaning

```python
## Missing values
dataset[dataset.isnull().any(axis = 1)]
```

```python
dataset.loc[:122, "Region"] = 0
dataset.loc[122:, "Region"] = 1
df = dataset
```

```python
df.info()
```

```python
df[['Region']] = df[['Region']].astype(int)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
## Removing the NULL values
df = df.dropna().reset_index(drop=True)
df.head()
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]]
```

```python
## Remove the 122nd row
df = df.drop(122).reset_index(drop=True)
```

```python
df.iloc[[122]]
```

```python
df.columns
```

```python
##Removing the spaces in the column names like the classes has a extra space
df.columns = df.columns.str.strip()
df.columns
```

```python
df.info()
```

## Changing the required columns as integrer data type

```python
df[['month','day','year','Temperature','RH','Ws']]=df[['month','day','year','Temperature','RH','Ws']].astype(int)
```

```python
df.info()
```

```python
### Changing the other columns to float datatypes
```

```python
objects = [features for features in df.columns if df[features].dtypes=='O']
```

```python
for i in objects:
    if i!='Classes':
        df[i]=df[i].astype(float)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.head()
```

```python
## Lets save the cleaned dataset
df.to_csv('Algerian_forest_fires_cleaned_dataset.csv', index=False)
```

## Exploratory Data Analysis

```python
## Drop day, month and year as it is not required
df_copy = df.drop(['day','month','year'], axis=1)
```

```python
df_copy.head()
```

```python
df_copy['Classes'].value_counts()
```

```python
df_copy['Classes'] = df_copy['Classes'].astype(str)
```

```python
## Encoding of the categories in classes
df_copy['Classes']= np.where(df_copy['Classes'].str.contains('not fire'),0,1)
```

```python
df_copy.head()
```

```python
df_copy['Classes'].value_counts()
```

```python
### Plot density plot for all features
df_copy.hist(bins=50, figsize=(20, 15))
sns.set(style="darkgrid")
plt.show()
```

```python
## Percentage for PIE-CHART
percentage = df_copy['Classes'].value_counts(normalize=True)
classlabels = ['Not Fire', 'Fire']
```

```python
## Plotting pie chart
classlabels= ['Fire', 'Not Fire']
plt.figure(figsize=(12,7))
plt.pie(percentage,labels=classlabels,autopct='%1.1f%%')
plt.title('Pie Chart of Classes')
plt.show()
```
